```python
import pandas as pd
```

```python
df=pd.read_csv('concrete.csv')
```

```python
df.head()
```

```python
df.head()
```

```python
x=df.drop('strength',axis=1)
y=df['strength']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=1)
x_train['strength']=y_train
```

```python
x_train.to_csv('concrete_train.csv',index=0)
x_test.to_csv('concrete_test.csv',index=0)
```
